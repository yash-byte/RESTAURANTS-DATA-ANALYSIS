User Rating of a restaurant plays a crucial role in selecting a restaurant or ordering the food from the restaurant.

Number of Votes given Restaurant

```python
data=pd.read_csv('zomato.csv',encoding='latin-1')
df=data.copy()
df=df[df['Country Code']==1]
df.dropna(inplace=True)
votes=np.array(df['Votes'])
rating=np.array(df['Aggregate rating'])
plt.scatter(votes,rating)
plt.xlabel('Votes', fontsize=16)
plt.title('Aggregate rating vs Votes',size=16)
plt.ylabel('Aggregate Rating', fontsize=16)
plt.show()
```

It is very evident from the plot that there is no direct relationship between no of votes and average rating so fact that more no of votes will lead to a good rating
is generally not true

```python
no_of_cuisines=[]
for i in df.index:
    l=len(df['Cuisines'][i].split(','))
    no_of_cuisines.append(l)
rating=np.array(df['Aggregate rating'])
plt.scatter(no_of_cuisines,rating)
plt.xlabel('No Of Cuisines', fontsize=16)
plt.ylabel('Aggregate Rating', fontsize=16)
plt.title('Aggregate rating vs No of Cuisines',size=16)
plt.tick_params(labelsize=15)
plt.show()
```

The Graph indicates to a bit extent that the restaurants
serving less number of cuisines have better rating
than the other ones seving 4 or more than 4
cuisines.Hence restaurats can be focused to practise
one cuisine effecitly for better rating

```python
average_cost=np.array(df['Average Cost for two'])
rating=np.array(df['Aggregate rating'])
plt.scatter(average_cost,rating)
plt.xlabel('Average cost for Two', fontsize=16)
plt.ylabel('Aggregate Rating', fontsize=16)
plt.title('Aggregate rating vs Average Cost For Two',size=16)
plt.xticks(rotation=90)
plt.xlim(0,8000)
plt.tick_params(labelsize=15)
plt.show()
```

it is not possible to establish a direct relationship between average cost for 2 and average rating but for some intervals  a trend can be observed

```python
df=data.copy()
df=df[df['Country Code']==1]
df.dropna(inplace=True)
d={}
for j in df.index:
    ele=data['Cuisines'][j]
    li=ele.strip().split(',')
    for v in li:
        d[v.strip()]=d.get(v.strip(),0)+1
count=Counter(d)
high=count.most_common(15)
print('---------------------------------------------------------------------------------------------------------------')
print('TOP 15 CUISINES SERVED IN MOST RESTAURANTS OF INDIA ')
cuisines=[]
for i in range(15):
    cuisines.append(high[i][0])
    print(i+1,':',high[i][0])
print('---------------------------------------------------------------------------------------------------------------')
cu=[]
rating=[]
for j in cuisines:
    ra=0
    l=0
    for i in df.index:
        ele=data['Cuisines'][i]
        li=ele.strip().split(',')
        if j in li:
            l+=1
            ra+=df['Aggregate rating'][i]
    cu.append(j)
    rating.append(ra/l)
plt.subplots(figsize=(20,10))
plt.title('RATING VERSES CUISINES SERVED IN INDIA')
plt.xlabel('Cuisine served')
plt.ylabel('Rating')
plt.plot(cu,rating,color='red')
plt.show()
```

In this plot i have taken top 15 cuisines sold in India and taken average rating for each and from this it can be observed that
to an extent cuisine plays a role in aggregate rating of Restaurant.Restaurants serving Continental have the highest average rating 
wheras Restaurants serving Mithai have the lowest averge rating

Find the weighted restaurant rating of each locality and find out the top 10 localities with more weighted restaurant rating?
Weighted Restaurant Rating= (number of votes * rating) /  (number of votes) .

```python
w_rating={}
for i in df.Locality.value_counts().index:
    votes=np.array(df[df.Locality==i]['Votes'])
    rating=np.array(df[df.Locality==i]['Aggregate rating'])
    c=np.array(votes*rating)
    if(votes.sum()!=0):
        val=c.sum()/votes.sum()
        w_rating[i]=val
w_rating=pd.Series(w_rating)
dict={'Locality' : w_rating.index,'Weighted Rating': w_rating.values}
w_rat=pd.DataFrame(dict).nlargest(10,'Weighted Rating')
w_rat.reset_index(drop=True,inplace=True)
w_rat.index+=1
print('---------------------------------------------------------------------------------------------')
print('                          TOP 10 LOCALITIES HAVING BEST RATING                               ')
print()
print('---------------------------------------------------------------------------------------------')
plt.subplots(figsize=(30,10))
plt.plot(w_rat['Locality'],w_rat['Weighted Rating'])
plt.ylabel('Weighted_Rating',size=13)
plt.xlabel('Locality',size=13)
plt.xticks(rotation=45)
plt.show()
w_rat
```
